```python
#creo una matrice 
diag1 = numpy.array( 
    [[1, 0],
     [0, 1]]
)
print("La taglia della matrice è: ", diag1.shape) 
print("Il numero di elementi della matrice è:", diag1.size)
```

```python
#creo una matrice diagonale e alla funzione "diag()" passo i valori che
# voglio dare agli elementi nella diagonale principale. Il numero di valori che metto
# sarà la taglia della matrice

matrice_1 = numpy.diag((1,1,1))
print ("La matrice ha dimensione: ", matrice_1.shape)
print (matrice_1)

#creo una  matrice identità. Alla funione identity() passo la taglia della matrice
identita_1 = numpy.identity(4)
print ("La matrice identità creata, ha dimensione: ", identita_1.shape)
print(identita_1)
```
